```python
import pandas as pd
```

```python
df = pd.read_csv('mahasiswa.csv')
df
```

```python
df.columns = ['No', 'NIM', 'Nama', 'JumlahSKS', 'NilaiTBQ']
df
```

```python
for index, row in df.iterrows():
    if row['JumlahSKS'] >= 110:
        df.loc[index, 'StatusKKN'] = 'Mengikuti'
    else:
        df.loc[index, 'StatusKKN'] = 'Tidak Boleh Mengikuti'
df
```

```python
for index, row in df.iterrows():
    if row['NilaiTBQ'] >= 70:
        df.loc[index, 'StatusTBQ'] = 'Lulus'
    else:
        df.loc[index, 'StatusTBQ'] = 'Tidak Lulus'
df
```

```python
dummies = pd.get_dummies(df.StatusKKN)
dummies
```

```python
dummies = pd.get_dummies(df.StatusTBQ)
dummies
```
