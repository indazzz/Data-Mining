```python
import pandas as pd
```

```python
df = pd.read_csv('covid.csv')
df
```

# penamaan variabel

```python
df.columns = ['type', 'f_type', 'id', 'p_code', 'p_name', 'confim_cases', 'recov_cases', 'death_case', 'FGT', 'FGC']
df
```

# cek data kosong

```python
df.isna().sum()
```

# cek tipe data

```python
df.dtypes
```

# buat variabel baru sesuai dengan kondisi

```python
for index, row in df.iterrows():
    if row['confim_cases'] >= 10000:
        df.loc[index, 'Status'] = 'Tinggi'
    else:
        df.loc[index, 'Status'] = 'Rendah'
df[['confim_cases', 'Status']]
```

# dummies

```python
dummies = pd.get_dummies(df.Status)
dummies
```

# rata-rata

```python
mean_confim_cases = df.confim_cases.mean()
mean_confim_cases
```

```python
mean_recov_cases = df.recov_cases.mean()
mean_recov_cases
```

```python
mean_death_case = df.death_case.mean()
mean_death_case
```

# minimal dan maksimal

```python
import statistics
from statistics import stdev
```

```python
std_confim_cases = (statistics.stdev(df.confim_cases))
std_confim_cases
```

```python
T_confim_cases_max = mean_confim_cases + (2*std_confim_cases)
T_confim_cases_max
```

```python
T_confim_cases_min = mean_confim_cases - (2*std_confim_cases)
T_confim_cases_min
```
