```python
import pandas as pd
```

```python
df = pd.read_csv('siswa.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df['UAS'] = df['UAS'].str.replace(",",".").astype(float)
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Tahun_lahir'] = row['Tgl Lahir'][-2:]
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Tahun']= + row['Tahun_Lahir']
df
```

```python
from datetime import datetime
now = datetime.now()

now.year
for index, row in df.iterrows():
    df.loc[index, 'Umur']=(now.year - row[' lulus']) + row['usia msk']
df
```

```python
d_train = df[['no Induk','Nama','UAS']].reset_index(drop=True)
d_train
```

```python
#import library yang akan digunakan
from IPython.display import Image
from sklearn.impute import SimpleImputer
from sklearn import tree
from subprocess import check_call
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
for index, row in d_train.iterrows():
    if row['UAS'] >=20:
        d_train.loc[index, 'StatusUAS'] = 'tinggi'
    else:
        d_train.loc[index, 'StatusUAS'] = 'rendah'
d_train
```

```python
y = d_train['StatusUAS'].values
y
```

```python
columns = ['no Induk', 'UAS']
features = d_train[list(columns)].values
features
```

```python
imputer = SimpleImputer(missing_values= np.NAN, strategy= 'mean', fill_value=None, verbose=0, copy=True)
x = imputer.fit_transform(features)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(x_train, y_train)
predict = clf.predict(x_test)
```

```python
predict, y_test
```

```python
with open("graph_tree.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names = columns, filled=True, rounded=True, special_characters = True)
```

```python
check_call(['dot', '-Tpng', 'graph_tree.dot', '-o', 'graph_tree.png'])
```

```python
Image(filename='graph_tree.png')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(5,3))
sns.heatmap(confusion_matrix(y_test, predict), annot=True, fmt='.0f', ax=ax)
plt.xlabel("predict")
plt.ylabel("actual")
plt.show
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict)
```
