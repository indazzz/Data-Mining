```python
import pandas as pd
```

```python
df = pd.read_csv('siswa.csv')
df
```

```python
df.dtypes
```

```python
df['UAS'] = df['UAS'].str.replace(",",".").astype(float)
df
```

```python
df.isna().sum()
```

```python
mean_UAS = df.UAS.mean()
mean_UAS
```

```python
import statistics
from statistics import stdev
```

```python
std_UAS = (statistics.stdev(df.UAS))
std_UAS
```

```python
T_UAS_max = mean_UAS + (2*std_UAS)
T_UAS_max
```

```python
T_UAS_min = mean_UAS - (2*std_UAS)
T_UAS_min
```
