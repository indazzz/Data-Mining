### Import Library

```python
import pandas as pd
```

### Import data tipe csv

```python
df = pd.read_csv('Data Alumni.csv')
df
```

### Cek tipe data per baris

```python
df.dtypes
```

```python
df.columns = ['No','NIM','Nama','TTL','Fakultas','Alamat','Masuk','Lulus','Status','MasaTunggu','Gaji','TOEFL','IPK']
df['IPK'] = df['IPK'].str.replace(",",".").astype(float)
df
```

```python
df.isna().sum()
```

### Mean Value

```python
mean_IPK = df.IPK.mean()
mean_TOEFL = df.TOEFL.mean()
mean_Masa = df.MasaTunggu.mean()
```

```python
mean_IPK
```

```python
mean_TOEFL
```

```python
mean_Masa
```

```python
df.IPK.fillna(mean_IPK, inplace=True)
df.TOEFL.fillna(mean_TOEFL, inplace=True)
df.MasaTunggu.fillna(mean_Masa, inplace=True)
df
```

```python
def_status = "Mahasiswa"
df.Status.fillna(def_status, inplace =True)
df
```

```python
df.isna().sum()
```

### Don't Care Method

```python
def_alamat = "-"
df.Alamat.fillna(def_alamat, inplace =True)
def_gaji = "0"
df.Gaji.fillna(def_gaji, inplace =True)
df
```

```python
df.isna().sum()
```

### Detecting Outlier

```python
import statistics
from statistics import stdev
```

```python
std_IPK = (statistics.stdev(df.IPK))
std_IPK
```

```python
T_IPK_max = mean_IPK + (2*std_IPK)
T_IPK_min = mean_IPK - (2*std_IPK)
```
