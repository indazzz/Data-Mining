```python
import pandas as pd
from openpyxl import load_workbook
from sklearn import preprocessing as pre
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv('Data_Alumni.csv')
df
```

```python
df.columns = ['No', 'NIM', 'Nama', 'TTL', 'Fakultas', 'Alamat', 'Masuk', 'Lulus', 'Status', 'MasaTunggu', 'Gaji', 'TOEFL', 'IPK']
```

```python
def_status = 'Mahasiswa'
df.Status.fillna(def_status, inplace = True)
```

```python
def_alamat = 'Kuningan'
df.Alamat.fillna(def_alamat, inplace = True)
```

```python
mean_gaji = df.Gaji.mean()
mean_TOEFL = df.TOEFL.mean()
mean_IPK = df.IPK.mean()
mean_masa = df.MasaTunggu.mean()
df.Gaji.fillna(mean_gaji, inplace = True)
df.TOEFL.fillna(mean_TOEFL, inplace = True)
df.IPK.fillna(mean_IPK, inplace = True)
df.MasaTunggu.fillna(mean_masa, inplace = True)
```

```python
print(df.dtypes)
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Masa Studi'] = row['Lulus'] - row['Masuk']
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Tahun Lahir'] = row['TTL'][-4:]
df
```

```python
from datetime import datetime
now = datetime.now()

now.year
for index, row in df.iterrows():
    df.loc[index, 'Umur'] = now.year - int (row['Tahun Lahir'])
    
df
```
