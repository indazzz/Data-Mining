```python
from openpyxl import load_workbook
import pandas as pd
```

```python
df = pd.read_csv ("Data Alumni.csv")
df
```

```python
df.loc[:, ['Nama','TTL']]
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Tahun Lahir'] = row['TTL'][-4:]
df
```

```python
from datetime import datetime
now = datetime.now()
now.year
for index, row in df.iterrows():
    df.loc[index, 'Umur']=now.year - int (row['Tahun Lahir'])
df
```

```python
for index, row in df.iterrows():
    df.loc[index, 'Masa Studi']=row['Tahun Lulus']-row['Tahun Masuk']
df
```
